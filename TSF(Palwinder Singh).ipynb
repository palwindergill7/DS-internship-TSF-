## Import Libraries

```python
import pandas as pp
import numpy as nn
import matplotlib.pyplot as ppt 
import seaborn as ss
```

```python
from sklearn.model_selection import train_test_split  
from sklearn.linear_model import LinearRegression
```

Importing Data

```python
gdata= pp.read_csv("http://bit.ly/w-data")
```

```python
gdata.head(5)
```

## Draw Plots Using dataset

Draw Scatter Plot Betweem Hours and Percentage 

```python
ppt.scatter('Hours','Scores',data=gdata,)
ppt.xlabel('Hours')
ppt.ylabel('Score')
ppt.title('Hours vs Scores in Pecentage')
```

Draw Regression Plot Betweem Hours and Percentage 

```python
ss.regplot('Hours','Scores',data=gdata)
ppt.title('Hours vs Percentage')
```

```python
X=gdata.iloc[:, :-1]
Y=gdata.iloc[:,1]
```

Split this data into training and test sets

```python
X_Train,X_Test,Y_Train,Y_Test= train_test_split(X,Y,test_size=0.3,random_state=0)
```

## Training the Data

```python
LR=LinearRegression()
LR.fit(X_Train,Y_Train)
```

Plot Regression using Straight line

```python
line= LR.coef_*X+ LR.intercept_
ppt.scatter(X,Y)
ppt.plot(X,line)
ppt.show()
```

## Predicting Values

```python
y_pred= LR.predict(X_Test)
```

```python
new_gdata=pp.DataFrame({'Actual':Y_Test,'Predicted':y_pred,'Difference':y_pred-Y_Test})
```
